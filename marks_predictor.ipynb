```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as lr
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("Data.csv")
df.head()
```

```python
x = df[["StudyHours"]]
y = df["Marks"]
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(x)
model = lr()
model.fit(X_poly,y);
```

```python
study_hours = float(input("Enter number of study hours:"))
study_df = pd.DataFrame([[study_hours]], columns=["StudyHours"])
study_poly = poly.transform(study_df)
prediction = model.predict(study_poly)[0]
prediction = max(0, min(100, prediction))
print(f"Predicted Score for {study_hours} study hours is: {int(prediction)}")
```
